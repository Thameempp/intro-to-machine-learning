# Bagging (Bootstrap Aggregating)

### Train multiple models in parallel on different random subsets of training data (with replacement).
* Combine predictions by majority voting for classification.
* Reduces variance and prevents overfitting.

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset and splitting 

x, y = load_iris(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    random_state=42
)
```

```python
model.fit(x_train, y_train)
print('Bagging Accuracy:', accuracy_score(y_test, model.predict(x_test)))
```
