### 1. Linear Algebra (Data Representation)
*Used to handle datasets as matrices and vectors.*
* **Vectors & Matrices:** Scalar vs. Vector vs. Matrix vs. Tensor.
* **Operations:** Dot products, Matrix multiplication, and Transposition.
* **Dimensionality Reduction:** Eigenvalues, Eigenvectors, and Singular Value Decomposition (SVD) for PCA.
* **Norms:** $L^1$ and $L^2$ norms (used in Lasso/Ridge regularization).

### 2. Calculus (Optimization)
*The engine that minimizes error (Loss Functions).*
* **Derivatives:** Calculating the slope to find the direction of steepest descent.
* **Partial Derivatives:** Used when the model has multiple features (parameters).
* **The Chain Rule:** The mathematical foundation of **Backpropagation** in Neural Networks.
* **Gradient Descent:** The algorithm used to update weights: $\theta = \theta - \alpha \nabla J(\theta)$.

### 3. Probability & Statistics (Uncertainty)
*Used to make predictions and validate model performance.*
* **Distributions:** Normal (Gaussian), Bernoulli, and Binomial.
* **Bayes’ Theorem:** Fundamental for Naive Bayes and probabilistic modeling.
* **Measures:** Mean, Median, Mode, Variance, and Standard Deviation.
* **Hypothesis Testing:** p-values and A/B testing to confirm if results are statistically significant.

### 4. Information Theory (Comparison)
*Used to measure the "distance" between predicted and actual values.*
* **Entropy:** Measuring the impurity of data (used in Decision Trees).
* **Cross-Entropy:** The standard loss function for classification tasks.
* **KL Divergence:** Measuring how one probability distribution differs from a second.
