# Feature Scaling

Feature scaling is the process of normalizing or standardizing numerical features in a dataset. Many machine learning algorithms are sensitive to feature scaling and perform better when features are on a similar scale.

## Why Scale Features?
- Algorithms like SVM, KNN, and Neural Networks are distance-based
- Prevents features with large ranges from dominating
- Improves convergence speed of gradient descent
- Helps with model interpretability

## Common Methods:
1. **Normalization** (Min-Max Scaling)
2. **Standardization** (Z-Score)
3. **Robust Scaling** (Resistant to outliers)
4. **Log Scaling** (For skewed distributions)

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/Users/thameem/Desktop/thameem/Machine Learning/Data/height-weight.csv')
print("Dataset shape:", df.shape)
print("\nFirst few rows:")
print(df.head())
print("\nDataset statistics before scaling:")
print(df.describe())
```

## 1. Min-Max Normalization

Scales values to a fixed range, typically [0, 1].

Formula: X_scaled = (X - X_min) / (X_max - X_min)

```python
# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax = pd.DataFrame(
    scaler_minmax.fit_transform(df),
    columns=df.columns
)
print("After Min-Max Scaling:")
print(df_minmax.describe())
print(f"Range: [{df_minmax.min().min()}, {df_minmax.max().max()}]")
```

## 2. Standardization (Z-Score Normalization)

Transforms data to have mean=0 and standard deviation=1.

Formula: X_scaled = (X - mean) / std

```python
# Standardization
scaler_standard = StandardScaler()
df_standard = df.copy()
df_standard = pd.DataFrame(
    scaler_standard.fit_transform(df),
    columns=df.columns
)
print("After Standardization:")
print(df_standard.describe())
print(f"Mean: {df_standard.mean().mean():.6f}")
print(f"Std: {df_standard.std().mean():.6f}")
```

## 3. Robust Scaling

Uses median and interquartile range, resistant to outliers.

```python
# Robust Scaling
scaler_robust = RobustScaler()
df_robust = df.copy()
df_robust = pd.DataFrame(
    scaler_robust.fit_transform(df),
    columns=df.columns
)
print("After Robust Scaling:")
print(df_robust.describe())

# Comparison
print("\n\nComparison of scaling methods:")
print("Original data range:")
for col in df.columns:
    print(f"{col}: [{df[col].min()}, {df[col].max()}]")
    
print("\nAfter Min-Max: ", f"[{df_minmax.min().min()}, {df_minmax.max().max()}]")
print("After Standardization mean/std: ", f"[{df_standard.mean().mean():.2f}, {df_standard.std().mean():.2f}]")
print("After Robust mean/std: ", f"[{df_robust.mean().mean():.2f}, {df_robust.std().mean():.2f}]")
```
