# Handling Missing Values

Missing values are a common problem in real-world datasets. They can arise from data collection errors, equipment failures, or incomplete data entry. Properly handling missing values is crucial for accurate model training.

## Types of Missing Data:
- **MCAR** (Missing Completely At Random) - no pattern
- **MAR** (Missing At Random) - pattern related to observed data
- **MNAR** (Missing Not At Random) - pattern related to unobserved data

## Strategies:
1. Deletion (Remove rows/columns)
2. Imputation (Fill with values)
3. Create indicator variables

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer, KNNImputer
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/Users/thameem/Desktop/thameem/Machine Learning/Data/Algerian_forest_fires_dataset_UPDATE.csv')
print("Dataset shape:", df.shape)
print("\nFirst few rows:")
print(df.head())

# Check for missing values
print("\nMissing values count:")
print(df.isnull().sum())
print("\nPercentage of missing values:")
print((df.isnull().sum() / len(df)) * 100)
```

## 1. Deletion Methods

Remove rows or columns with missing values. Use when missing data is minimal.

```python
# Method 1: Drop rows with any missing values
df_dropped_rows = df.dropna()
print(f"Shape after dropping rows with missing values: {df_dropped_rows.shape}")

# Method 2: Drop rows where all values are missing
df_dropped_all = df.dropna(how='all')
print(f"Shape after dropping rows with all missing values: {df_dropped_all.shape}")

# Method 3: Drop columns with missing values
df_dropped_cols = df.dropna(axis=1)
print(f"Shape after dropping columns with missing values: {df_dropped_cols.shape}")
```

## 2. Imputation Methods

Fill missing values with estimated values based on available data.

```python
# Method 1: Mean imputation
df_mean = df.copy()
numeric_cols = df_mean.select_dtypes(include=[np.number]).columns
imputer_mean = SimpleImputer(strategy='mean')
df_mean[numeric_cols] = imputer_mean.fit_transform(df_mean[numeric_cols])
print("Mean imputation - Missing values after:", df_mean.isnull().sum().sum())

# Method 2: Median imputation
df_median = df.copy()
imputer_median = SimpleImputer(strategy='median')
df_median[numeric_cols] = imputer_median.fit_transform(df_median[numeric_cols])
print("Median imputation - Missing values after:", df_median.isnull().sum().sum())

# Method 3: Mode imputation (for categorical)
df_mode = df.copy()
categorical_cols = df_mode.select_dtypes(include=['object']).columns
imputer_mode = SimpleImputer(strategy='most_frequent')
if len(categorical_cols) > 0:
    df_mode[categorical_cols] = imputer_mode.fit_transform(df_mode[categorical_cols])
print("Mode imputation - Missing values after:", df_mode.isnull().sum().sum())
```

## 3. Advanced Imputation: KNN Imputation

Use K-Nearest Neighbors to impute missing values based on similar samples.

```python
# KNN Imputation
df_knn = df.copy()
numeric_cols = df_knn.select_dtypes(include=[np.number]).columns
imputer_knn = KNNImputer(n_neighbors=5)
df_knn[numeric_cols] = imputer_knn.fit_transform(df_knn[numeric_cols])
print("KNN imputation - Missing values after:", df_knn.isnull().sum().sum())
print("\nComparison of imputation methods:")
print(f"Original missing values: {df.isnull().sum().sum()}")
print(f"After Mean imputation: {df_mean.isnull().sum().sum()}")
print(f"After KNN imputation: {df_knn.isnull().sum().sum()}")
```
