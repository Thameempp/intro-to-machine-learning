# Data Cleaning Techniques

Data cleaning is the process of identifying and correcting (or removing) errors and inconsistencies in datasets. It's a crucial step in the data preprocessing pipeline that ensures data quality and improves model performance.

## Key Objectives:
- Remove duplicates
- Fix inconsistent formatting
- Handle missing values
- Remove unnecessary columns
- Fix data type issues
- Standardize text fields

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load the messy dataset
df = pd.read_csv('/Users/thameem/Desktop/thameem/Machine Learning/Data/messy_data.csv')
print("Dataset shape:", df.shape)
print("\nFirst few rows:")
print(df.head())
print("\nDataset info:")
print(df.info())
print("\nDataset description:")
print(df.describe())
```

## 1. Removing Duplicates

Duplicate records can skew analysis and model results. Identify and remove them efficiently.

```python
# Check for duplicates
print("Total duplicate rows:", df.duplicated().sum())
print("Percentage of duplicates:", (df.duplicated().sum() / len(df)) * 100, "%")

# Remove duplicates
df_cleaned = df.drop_duplicates()
print(f"\nShape before: {df.shape}")
print(f"Shape after removing duplicates: {df_cleaned.shape}")
```

## 4. Standardizing Text Fields

Ensure consistency in text data (lowercase, remove whitespace, etc.)

```python
# Standardize text columns
text_columns = df_cleaned.select_dtypes(include=['object']).columns

for col in text_columns:
    if df_cleaned[col].dtype == 'object':
        # Convert to lowercase and strip whitespace
        df_cleaned[col] = df_cleaned[col].str.lower().str.strip()
        print(f"Standardized column: {col}")

print("\nCleaned dataset:")
print(df_cleaned.head())
```
