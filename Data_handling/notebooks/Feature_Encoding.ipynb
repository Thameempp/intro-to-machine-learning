# Feature Encoding

Feature encoding is the process of converting categorical variables into numerical format that machine learning algorithms can understand. Most ML algorithms require numerical inputs.

## Types of Categorical Variables:
- **Nominal** - No order (colors, animal types)
- **Ordinal** - Has order (education level, ratings)

## Encoding Methods:
1. **One-Hot Encoding** - For nominal variables
2. **Label Encoding** - For ordinal variables
3. **Binary Encoding** - For binary variables
4. **Target Encoding** - Using target variable relationship
5. **Frequency Encoding** - Based on frequency of categories

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder

# Create sample categorical data
df = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue', 'Green'],
    'Size': ['Small', 'Medium', 'Large', 'Small', 'Large', 'Medium'],
    'Quality': ['Good', 'Average', 'Excellent', 'Good', 'Good', 'Average']
})

print("Original Data:")
print(df)
print("\nData types:")
print(df.dtypes)
```

## 1. One-Hot Encoding

Creates binary columns for each category. Best for nominal variables with few categories.

```python
# One-Hot Encoding
df_onehot = pd.get_dummies(df, columns=['Color'])
print("One-Hot Encoded (Color):")
print(df_onehot)
```

## 2. Label Encoding

Converts categories to integer labels. Suitable for ordinal variables.

```python
# Label Encoding for ordinal data (Size: Small < Medium < Large)
df_label = df.copy()
le = LabelEncoder()

# Define order for Size
size_order = {'Small': 0, 'Medium': 1, 'Large': 2}
df_label['Size_encoded'] = df_label['Size'].map(size_order)

print("\nLabel Encoded (Size):")
print(df_label[['Size', 'Size_encoded']])
```

## 3. Binary Encoding

For binary categorical variables (Yes/No, True/False).

```python
# Binary Encoding
df_sample = pd.DataFrame({
    'IsPurchased': ['Yes', 'No', 'Yes', 'No', 'Yes']
})

df_binary = df_sample.copy()
df_binary['IsPurchased_encoded'] = (df_binary['IsPurchased'] == 'Yes').astype(int)

print("Binary Encoded:")
print(df_binary)
```

## 4. Frequency Encoding

Replaces categories with their frequency counts.

```python
# Frequency Encoding
df_freq = df.copy()
frequency_map = df_freq['Color'].value_counts().to_dict()
df_freq['Color_frequency'] = df_freq['Color'].map(frequency_map)

print("Frequency Encoded (Color):")
print(df_freq[['Color', 'Color_frequency']])
print(f"\nFrequency mapping: {frequency_map}")
```
