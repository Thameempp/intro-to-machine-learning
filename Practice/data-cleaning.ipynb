```python
# remaining task 

# fill the empty department column based on salary range and experience
'''
High salary + high experience → IT / Finance
Low salary + low experience → HR
Mid combos → mixed

'''
```

```python
df[df['department']==0]
```

```python
# Creating groups using BOTH features
df['salary_grp'] = pd.qcut(df['salary'], q=3, labels=['low', 'medium', 'high'])
df['experience_grp'] = pd.qcut(df['experience'], q=3, labels=['junior', 'mid_exp', 'senior'])
```

```python
#treating the zero in department as nan value

df['department'] = df['department'].replace(0, np.nan)
```

```python
#now filling the department column based on new columns salary_grp and exp_grp

df['department'] = df['department'].fillna(
    df.groupby(['salary_grp', 'experience_grp'])['department']
    .transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
)
```

```python
# now checking the row where department was zero 
df.iloc[42]
```

```python
# now droping the new columns 
df.drop(['salary_grp','experience_grp'], axis=1, inplace=True)
```

```python
# filling city with mode based on department

# finding how many city got missing here
df[df['city']==0]
```

```python
# treating zero as nan 

df['city'] = df['city'].replace(0, np.nan)
df['city'].head()
```

```python
# now filling with mode 
df['city'] = df['city'].fillna(
    df.groupby('department')['city']
    .transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
)
df['city'].head()
```
