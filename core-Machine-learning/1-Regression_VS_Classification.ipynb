# Regression vs classification :

## What is Regression:
- Supervised learning technique that fits a line or curve to data points to minimize prediction errors
### types of regressions:
- Linear: straight line : 
predicts continuous values using one feature and a straight line
- Multiple Linear: many features : 
extends this to multiple features
- Polynomial: curves : 
captures curves by adding powers
- Logistic: binary outcomes : 
Logistic regression models probabilities for classification (not true regression) outputting 0-1 values

## why do we need regression : 

- Regression helps us estimate or predict a numerical value using past data
- To Understand Relationships Between Variables
- Decision Making in Business

### examples:  

- What will next month’s sales be?
- What price should a house have?
- Pricing strategy , Budget planning
- Which factors matter most
- Whether impact is positive or negative

## what is classification: 
- the model learns to assign data into categories (classes or labels).

- Instead of predicting a number, classification answers:

    - Is this spam or not spam?

    - Is the patient sick or healthy?

## regression vs classification :

- regression - predicts numbers (output is number)
- classification - predicts categories or assign class (output is label)

- graph in regression - best fit-line
- graph in classification - decision boundary
- Both uses similar input , Difference is only target/output type. 

### examples:

- Email → Spam / Not Spam
- Disease → Yes / No
- Image → Cat / Dog
